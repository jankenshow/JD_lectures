```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation
```

```python
x = np.arange(-3, 3, 0.5)
y = np.arange(-3, 3, 0.5)
X, Y = np.meshgrid(x, y)
```

```python
a = 3
b = 2
c = 1
Z = - (a/c) * X - (b/c) * Y
```

```python
fig = plt.figure()
ax = Axes3D(fig)

def plane_3dplot(X, Y, a=1, b=1, c=1, option={}, fig=fig, ax=ax):

    Z = - (a/c) * X - (b/c) * Y

    ax.set_xlim(-10, 10)
    ax.set_ylim(-10, 10)
    ax.set_zlim(-10, 10)

    # ax.plot_wireframe(X,Y,Z)
    ax.scatter3D(np.ravel(X),np.ravel(Y),np.ravel(Z))
    ax.quiver(0, 0, 0, a, b, c, color='red')
    
    return fig, ax

def init():
    plane_3dplot(X, Y, a, b, c)
    return fig,

def animate(i):
    ax.view_init(elev=10., azim=i)
    return fig,

anim = animation.FuncAnimation(fig, animate, 
                               init_func=init,
                               frames=280, 
                               interval=20, 
                               blit=True)
anim.save('3d_animation.mp4',
          fps=30,
          extra_args=['-vcodec', 'libx264'])
```
