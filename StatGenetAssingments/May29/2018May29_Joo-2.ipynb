2.Definitions and mathematical preliminaries

2.1 The concept of graph
Graph G : a set of vertices V(G) and a set of edges E(G)
The number of vertices: |V(G)|
The set of vertices adjacent to a vertex v1: N(v1)
The degree of a vertex v1: the number of edges with which it is incident d(v1)
Subgraph G': all vertices and edges are included in G. G is supergraph of G'.

2.2 Directed and undirected graphs
Directed graph: no distinction between the two vertices associated with each edge.
Undirected graph: its edges are directed from one vertices to aother.

2.3 Node-degree and the adjacency matrix
・Undirected graph
d(u): the degree of node u in V(G) = the total number of edges at u = |N(u)|
・Directed graph
d+(u): the number of edges that terminate at u.
The adjacensy matrix A: 点iと点jを結ぶ辺の本数を第ij要素とするn × nの行列

2.4 Path, path length and connecter graph
path length: distance d(u,v)
connected graph: there is some path from u to v, for every pair of vertices,(u,v)
