複雑なネットワークは、コミュニケーションから分子生物学や集団生物学まで、幅広い学問において重要な役割を担っているが、計算生物学において、グラフ理論が注目を集めている。

グラフ理論は大規模なデータに寛容であり、論理的に理解がしやすく、さらに体系的なアプローチが可能であるため、複雑な生物学的ネットワークの数学的なモデリングが可能であり、例えば生体分子の相関ネットワークをモデルによって理解することができれば、ガンなどの病気に対する効果的な治療戦略を開発することができる。

論文内では、以下の分野における、グラフ理論の応用研究例の紹介がされている。
・タンパク同士の相互作用や代謝、転写制御などの生物分子ネットワークのモデリング
・全体としての生体分子ネットワークの中で、ある遺伝子やタンパクまたはそれらの相互作用の重要性を測定する
・複数の遺伝子やタンパクがある機能を発揮する(プロセスを行う)ための群を特定する
・生体分子のネットワークにおける一定の重要な経路(パスウェイ)を特定する(様々な種の生物に共通する経路、生物の類似性に大きく関わる経路など)

ネットワークレベルの膨大なデータセットに対応したグラフ理論を用いることで、生物学的なネットワークの数学的な数学的なモデリングを可能とし、病気の原因分子の特定など、現在の医療問題を解決することが可能となる。
