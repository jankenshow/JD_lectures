Summary of the document "Graph Theory and Analysis of Biological Data in Computational Biology"

全7章の構成になっている。第1章がイントロダクション。第2章がグラフ理論の定義と例。第3章から第6章では一つずつ大きな問題を取り上げて、基本的なコンセプト・近年の進展・今後の研究の可能性について述べている。第7章は結論。

第1章

様々な場面で全体の統制の様子を調べる際、複雑ネットワークによるアプローチが有用である。この文章では、計算生物学におけるグラフ理論の応用を扱う。

第2章

グラフの定義や例。グラフとは頂点と辺からなる図形。辺を繋がりと見なして、隣り合うということが定義できる。辺には向きをつけることもできる。この繋がりを数値化して行列を作ることで、繋がりの様子を計算することができる。

第3章

分子ネットワークのモデリングについて。転写規則、たんぱく質相互作用、代謝ネットワークという3つの例について、グラフを用いたモデル化をそれぞれどう行うかを述べている。遺伝子やたんぱく質が頂点、作用が辺で表現される。作用の様子を反映できるよう、辺に向きをつける、辺の様子によって頂点をグループ化するといった実装を行う。ネットワークが時間依存で変化したり、複数のインプットから複数のアウトプットが得られるような反応は、グラフによる表現が難しい。

第4章

分子ネットワーク内で何が重要か。単一の遺伝子（＝頂点）のみに注目するのではなく、小さなブロック（＝決まった繋がり方をしたいくつかの頂点）を考え、モジュールと呼ぶ。こういった小さなブロックの役割やその重要性を理解することが大事。計算機を用いて、このようなアプローチが成功している例が多くある。

第5章

生体ネットワークにおけるモジュールの同定について。第4章と同様に、小さなブロック単位で役割を理解するというアプローチ。モチーフと呼ばれる非常に小さなブロック（頂点数3～4程度）が注目されている。その理由は、生体ネットワークにおける出現頻度が非常に高いこと、トポロジカルな解析によって同定することが容易なこと。これに関連する結果も多くあり、解析の有力候補と見なされている。

第6章

分子ネットワークにおける新たなパスウェイの発見。代表的な問題がいくつかあるが、核として共通しているのは、異なるネットワークにおいてどんなパスウェイを同じと見なすかということ。複雑さや煩雑さから、実験で解明していくのは困難な段階になっている。計算機が自動的に推定してくれるところまでは進展していないが、これまでの知見を活かした計算機によるアプローチが進められている。いずれ、合成過程への預言がなされることが期待されている。

第7章

生物学において、ネットワークはどこにでも現れる。また、それは大規模なデータになりやすい。グラフ理論はそれに対するアプローチを可能にする。今日のグラフ理論では、生物学における反応の複雑さを表現しきれていない部分もあるが、理論の進展によってこの問題は解決できるものと思われる。こういったアプローチから、例えばガンに対する戦略的な対処方法が解明されていくことが期待されている。
